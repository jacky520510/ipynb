```python
need_feature=['ESAL','DISTRICT_CITY_CODE','SURFACE_LAYER_CODE','BOTTOM_LAYER_CODE']
need_df=df[need_feature]
need_df
```

```python
import matplotlib.pyplot as plt
def plot_data(titie,test,predict):
    plt.figure(figsize=(8,6),dpi=100)
    plt.plot(range(len(test)),test,label='y_true')
    plt.plot(range(len(test)),predict,label='predict')
    plt.title(titie)
    plt.legend()
    plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
Y=['PCI','RQI','RDI']

def predice_model(y):
    model=RandomForestRegressor()
    df_y=df[y]
    df_X=need_df.copy()
    ##划分测试和训练
    x_train,x_test,y_train,y_test=train_test_split(df_X,df_y,random_state=233)
    model.fit(x_train,y_train)
    predict=model.predict(x_test)
    mse=mean_squared_error(predict,y_test)
    print(f"{y} predict MSE:{mse}")
    plot_data(y,y_test,predict)
    return model
```

```python
pci_model=predice_model(Y[0])
```

```python
rqi_model=predice_model(Y[1])
```

```python
rdi_model=predice_model(Y[2])
```

```python
import numpy as np
input_data=[[1200,500000,2.0,1]]
pci=pci_model.predict(np.array(input_data).reshape(1,4))
rqi=rqi_model.predict(np.array(input_data).reshape(1,4))
rdi=rdi_model.predict(np.array(input_data).reshape(1,4))
print("PCI:",pci[0],"RQI:",rqi[0],"RDI:",rdi[0])
```
