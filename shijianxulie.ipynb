# 指数平滑法
仅能预测下一个时间

```python
表格4=pd.read_excel('表格004.xlsx',)
表格4.drop(['Unnamed: 0'],inplace=True,axis=1)
表格4.head()
```

```python
import numpy as np
# 单指数平滑
def exponential_smoothing(series, alpha):
    """
        series - dataset with timestamps
        alpha - float [0.0, 1.0], smoothing parameter
    """
    result = np.array([series[0]]) # first value is same as series
    for n in range(1, len(series)):
        result = np.append(result,alpha * series[n] + (1 - alpha) * result[n-1])
    return result
#前一个是较新的时间
def predict_next_time(y_1,y_2,alpha):
    return alpha * y_1 + (1 - alpha) * y_2

def predict(series,alpha):
    smooth_series = exponential_smoothing(series,alpha)
    predict_value = predict_next_time(series[-1],smooth_series[-1],alpha)
    smooth_series = np.append(smooth_series,predict_value)
    return smooth_series

#计算MSE
def mse(first_series,second_series):
    mse = (first_series-second_series)**2/len(first_series)
    return mse
```

```python
data = 表格4[表格4.ROAD_SECTION_CODE=='CQNHNZ0000-0100-1']
data = data.sort_values(by='duration')
data
```

## 对数据分类型，根据mse计算alpha参数

```python
predict(data['PCI'][:4].values,0.3)
```

```python
data_by_cat = {}
data_by_code={}
for name,group in 表格4.groupby('类型'):
    data_by_cat[name] = group
for name,group in data_by_cat['类型三'].groupby('ROAD_SECTION_CODE'):
    data_by_code[name]=group.sort_values(by='duration')
```

```python
#搜索最佳的alpha参数
mmse = np.inf
best_alpha=0
for alpha in np.arange(0,1,0.05):
    mse_list = np.array([])
    for value in data_by_code.values():
        predict_list=predict(value['PCI'].values[:4],alpha)
        _mse = mse(value['PCI'].values[-1:],predict_list[-1:])
        mse_list = np.append(mse_list,_mse)
        mse_list_mean = mse_list.mean()
    if mse_list_mean<mmse:
        mmse = mse_list_mean
        best_alpha = alpha
print(best_alpha)
```
